```python
#A simple Formula to find Your Amounnt: Compound Interest 
P = float(input("What is your Principal: "))
R = float(input("What is your rate: "))
t = float(input("What is yout time: "))
n = float(input("Number of times interest is compounded per year: "))
#Displaying the amount when calculating compound interest
A = P*(1 + R/n)**(n*t)
print("The compound amount is:", A)
print(type(A))
```

```python
#A simple Formula to find your Amount: Simple Interest 
P = float(input("What is your Principal: "))
R = float(input("What is your rate: "))
T = float(input("What is yout time: "))

#Displaying the amount when calculating Simple interest
A = P*(1 + R/100*T)
print("The Simple Interest amount is:", A)
print(type(A))
```

```python
# A simple formula to find the value for your Annuity Plan

PMT = float(input("how much you deposit each period: "))
R = float(input("What is your Annual interest rate: "))
n = float(input("Number of times interest is compounded per year: "))
t = float(input(" what is your time: "))

#Displaying your value for your Annuity plan 
A = PMT*((1 + R/n)**(n*t) - 1)/(R/n)
print("Here is your value for your Annuity plan",A)
```
